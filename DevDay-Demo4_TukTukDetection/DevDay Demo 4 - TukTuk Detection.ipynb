Tuk Tuk Detection using Trained Model 

```python
from imageai.Detection.Custom import CustomObjectDetection
```

Load Trained model and other params 

```python
detector = CustomObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath("detection_model-ex-010--loss-0006.270.h5") 
detector.setJsonPath("detection_config.json")
detector.loadModel()
```

Perform the TukTuk Detection 

```python
detections = detector.detectObjectsFromImage(input_image="test3.jpg", output_image_path="tuktuk-detected.jpg")
for detection in detections:
    print(detection["name"], " : ", detection["percentage_probability"], " : ", detection["box_points"])
```
